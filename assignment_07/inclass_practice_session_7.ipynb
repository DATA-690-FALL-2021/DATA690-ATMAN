```python
df  = pd.read_csv('wdi_data.csv')
df.head()
```

```python
# Drop a column

df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.head()
```

```python
# df_ind = df_new.query("Country Code == 'IND'")

df_ind = df[df["Country Code"].isin(['IND','BRA'])]
df_ind
```

```python
df_ind_gdppcap = df_ind.query("indicator == 'NY.GDP.PCAP.PP.CD'")
df_ind_gdppcap
```

```python
fig = go.Figure()

trace_0=go.Bar(
    x = df_ind_gdppcap["Year"],
    y = df_ind_gdppcap["value"]
)

fig.add_trace(trace_0)

fig.update_layout(
    title="India GDP per Capita over Time",
    xaxis_title="YEAR",
    yaxis_title="GDP per Capita"
)

fig.show()
```

```python
fig = px.bar(
    data_frame = df_ind_gdppcap,
    title="India GDP per Capita over Time",
    labels={"value" : "GDP per Capita", "Year":"YEAR"},

   # Horizontal or Vertical representation
    orientation="v",
    x = "Year",
    y = "value",
    color = "Country Code",
    barmode = "group"
)

fig.update_xaxes(tickangle=45)

fig.show()
```

```python
df_ind_bra_gdppcap = df_ind.query("indicator == 'NY.GDP.PCAP.PP.CD'")
df_ind_bra_gdppcap
```

```python
fig = go.Figure()

df_ind = df_ind_bra_gdppcap[df_ind_bra_gdppcap["Country Code"]=="IND"]
trace_0=go.Bar(
    x = df_ind["Year"],
    y = df_ind["value"],
    name = "India"
)
df_bra = df_ind_bra_gdppcap[df_ind_bra_gdppcap["Country Code"]=="BRA"]
trace_1=go.Bar(
    x=df_bra["Year"],
    y=df_bra["value"],
    name = "Brazil"
)

fig.add_trace(trace_0)
fig.add_trace(trace_1)
fig.update_layout(
    title="India vs Brazil GDP per Capita over Time",
    xaxis_title="YEAR",
    yaxis_title="GDP per Capita"
)

fig.show()
```
