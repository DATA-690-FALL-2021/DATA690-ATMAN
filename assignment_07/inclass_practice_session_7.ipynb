```python
import pandas as pd
import numpy as np
import plotly as pt
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
print("Pandas Version is:", pd.__version__)
print('\n')

print("Plotly Version is:", pt.__version__)
print('\n')

print("Numpy Version is:", np.__version__)
print('\n')
```

```python
df  = pd.read_csv('wdi_data.csv')
df.head()
```

```python
# Drop a column

df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.head()
```

```python
# df_ind = df_new.query("Country Code == 'IND'")

df_ind = df[df["Country Code"]=='IND']
df_ind
```

```python
df_ind_gdppcap = df_ind.query("indicator == 'NY.GDP.PCAP.PP.CD'")
df_ind_gdppcap
```

```python
fig = go.Figure()

trace_0=go.Bar(
    x = df_ind_gdppcap["Year"],
    y = df_ind_gdppcap["value"]
)

fig.add_trace(trace_0)

fig.update_layout(
    title="India GDP per Capita over Time",
    xaxis_title="YEAR",
    yaxis_title="GDP per Capita"
)

fig.show()
```

```python
fig = px.bar(
    data_frame = df_ind_gdppcap,
    title="India GDP per Capita over Time",
    labels={"value" : "GDP per Capita", "Year":"YEAR"},
    x = "Year",
    y = "value"
)

fig.show()
```
