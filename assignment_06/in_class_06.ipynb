```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("wdi_data.csv")
df.head()
```

```python
df.tail(5)
```

```python
df = df[df["indicator"]=='SP.POP.TOTL']
df = df[df["Country Code"]=='USA']

df.head(20)
```

```python
# BAR PLOT

fig =go.Figure()

trace_0=go.Bar(
    x = df["Year"],
    y = df["value"]
)

fig.update_layout(
    title = "USA POPULATION GROWTH",
    xaxis={"title":"Year"},
    yaxis={"title":"Population in millions"}
    
)

fig.add_trace(trace_0)
fig.show()
```

```python
# LINE PLOT

fig =go.Figure()

trace_0=go.Line(
    x = df["Year"],
    y = df["value"]
)

fig.update_layout(
    title = "USA POPULATION GROWTH",
    xaxis={"title":"Year"},
    yaxis={"title":"Population in millions"}
    
)

fig.add_trace(trace_0)
fig.show()
```

```python
# SCATTER PLOT

fig =go.Figure()

trace_0=go.Scatter(
    x = df["Year"],
    y = df["value"],
    mode = "markers"
)

fig.update_layout(
    title = "USA POPULATION GROWTH",
    xaxis={"title":"Year"},
    yaxis={"title":"Population in millions"}
    
)

fig.add_trace(trace_0)
fig.show()
```
