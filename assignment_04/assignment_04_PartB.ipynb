##6. Cleanse the data to make the year, population, cost, cost per person numerical type

```python
def cleanse(a_string):
  return int(a_string.replace(",","").replace("*",""))
df["year_list"] = df["Census Year"].apply(cleanse)  
df["tot_pop"] = df["Total Population"].apply(cleanse)
df
```
